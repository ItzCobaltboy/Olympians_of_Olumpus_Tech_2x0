```python
#code for creating a bar plot for quality index vs no of wines
#including standard libraries..
import pandas as pd 
import matplotlib.pyplot as plt

#reading the csv file and declaring the data frame.
wine_details = pd.read_csv('PS_dataset (1).csv')

#counting occurance of each quality score,this returns no of wines with each quality score with index being the quality score and values are no of wines.
# [quality_count.index]->quality score,[quality_count.values]->no of wines with that quality score.
quality_count = wine_details['quality'].value_counts().sort_index()

#declaring barplot size
plt.figure(figsize=(8,6))

#plotting the bar graph
plt.bar(quality_count.index,quality_count.values,color='blue')
plt.xlabel('Quality score')
plt.ylabel('No of Wines')
plt.title('Count of wines for each quality score.')
plt.xticks(quality_count.index)
plt.show()
```

```python
#code for  Creating a scatter plot of alcohol vs. pH.
import pandas as pd 
import matplotlib.pyplot as plt

#reading the csv file and declaring the data frame.
wine_details = pd.read_csv('PS_dataset (1).csv')

# Scatter plot for alcohol vs. pH,'alpha' makes the points plotted transluscent.to see overlapping.
plt.figure(figsize=(8, 6))
plt.scatter(wine_details['alcohol'], wine_details['pH'], alpha=0.5, color='red')
plt.xlabel('Alcohol')
plt.ylabel('pH')
plt.title('Scatter Plot of Alcohol vs. pH')
plt.show()
#there seem to be coorelation from the graph.
```

```python
#code for creating a histogram of alcohol to visualize its distribution.
import pandas as pd 
import matplotlib.pyplot as plt

#reading the csv file and declaring the data frame.
wine_details = pd.read_csv('PS_dataset (1).csv')


#Histogram for alcohol vs no of wines correspoding to it
plt.figure(figsize=(10, 6))
plt.hist(wine_details['alcohol'],bins=30,color='blue',edgecolor='black')
plt.xlabel('Alcohol')
plt.ylabel('No of Wine')
plt.title('Graph for alcohol distribution')
plt.show()
```

```python
#code for creating a boxplot to compare the spread of residual sugar values grouped by quality score.
import pandas as pd 
import matplotlib.pyplot as plt

#reading the csv file and declaring the data frame.
wine_details = pd.read_csv('PS_dataset (1).csv')

#creating a box plot
plt.figure(figsize=(10, 6))
wine_details.boxplot(column='residual sugar', by='quality', grid=False)
plt.title('Residual Sugar by Quality Score')
#plt.suptitle('')  # Suppresses the automatic 'by' title
plt.xlabel('Quality Score')
plt.ylabel('Residual Sugar (g/L)')
plt.show()
```

```python
# code for  Plotting a correlation matrix for the entire dataset and use a heatmap to visualize it.
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

#reading the csv file and declaring the data frame.
wine_details = pd.read_csv('PS_dataset (1).csv')

# Calculate the correlation matrix
correlation_matrix =wine_details.corr()

#plotting correlation matrix,heatmap.
plt.figure(figsize=(12, 8))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar(label='Correlation Coefficient')
plt.title('Correlation Matrix Heatmap')

# Adding labels
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45, ha='right')
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)

# Displaying values inside the heatmap cells
for i in range(len(correlation_matrix.columns)):
    for j in range(len(correlation_matrix.columns)):
        plt.text(j, i, f"{correlation_matrix.iloc[i, j]:.2f}", ha='center', va='center', color='black')

plt.tight_layout()
plt.show()
```

```python
#code for  creating a line plot showing the trend of total sulfur dioxide against free sulfur dioxide.
import pandas as pd
import matplotlib.pyplot as plt

#reading the csv file and declaring the data frame.
wine_details = pd.read_csv('PS_dataset (1).csv')


#plotting line graph
plt.figure(figsize=(10, 6))
plt.plot(wine_details['free sulfur dioxide'], wine_details['total sulfur dioxide'], color='blue',markeredgecolor='red', marker='o', linestyle='-')
plt.title('Trend of Total Sulfur Dioxide vs Free Sulfur Dioxide')
plt.xlabel('Free Sulfur Dioxide (mg/L)')
plt.ylabel('Total Sulfur Dioxide (mg/L)')
plt.grid(True)
plt.show()
```
